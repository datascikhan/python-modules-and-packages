
# Step 2: Import the Module

Python me **module import karne** ka use hum apne program me **dusre file ke functions, classes, aur variables** use karne ke liye karte hain.  
- `import module_name` → pura module import hota hai, fir module ke components ko `module_name.component` ke through access karte hain.  
- `from module_name import component` → specific function, class ya variable directly import hota hai.  
- `import module_name as alias` → module ko short alias name ke saath import kar sakte hain.

# 3️⃣ Import Variations

| Import Method   | Example                   | Description                             |
| --------------- | ------------------------- | --------------------------------------- |
| Standard import | `import module`           | Full module access via `module.name`    |
| Specific import | `from module import func` | Direct access to function or variable   |
| Import all      | `from module import *`    | Access all functions/variables directly |
| Alias           | `import module as m`      | Shorter name for module                 |
