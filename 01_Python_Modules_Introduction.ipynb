# Python Modules 
 
Python me **modules** ka use code ko **organize aur reuse** karne ke liye hota hai.  
- Module ek **Python file (.py)** hoti hai jisme functions, classes, variables ya runnable code define kiya ja sakta hai.  
- Modules ke zariye hum **code ko multiple files me divide** karke **maintainable aur reusable** bana sakte hain.  
- Python me built-in modules bhi available hain (jaise `math`, `os`, `random`) aur hum **custom modules** bhi create kar sakte hain.  
- Module ko import kar ke hum uske content ko apne program me use kar sakte hain using **`import`** keyword.

**Benefits:**  
- Code reusability  
- Maintainability  
- Namespace separation  
- Collaboration in large projects


# 1️⃣ What is a Module?


Python me **module** ek **Python file (.py)** hoti hai jisme hum **functions, classes, variables** aur **executable code** define kar sakte hain.  
Modules ka main purpose hai **code ko organize aur reuse** karna.  

- **Reusability:** Ek module me likha code multiple programs me use kiya ja sakta hai.  
- **Organization:** Complex programs ko chhote, manageable files me divide karna easy hota hai.  
- **Namespace management:** Modules alag namespace provide karte hain, jisse variable name conflicts avoid hote hain.  

**Built-in Modules:**  
Python ke saath kai ready-made modules aate hain jaise:  
- `math` → mathematical operations  
- `os` → operating system related operations  
- `random` → random numbers generate karna  

**Custom Modules:**  
Hum apni Python file ko module bana kar uske functions ya classes ko **import** karke use kar sakte hain.  

**Example:**  

import math

print(math.sqrt(16))  # Using sqrt() function from math module
