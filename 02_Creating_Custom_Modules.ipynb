# 2️⃣ Creating a Custom Module
 
Python me **custom module** create karne ke liye sabse pehle ek **Python file (.py)** banate hain jisme hum apne **functions, classes, aur variables** define karte hain.  
- Ye module **reusable code** provide karta hai.  
- Module ko **import** karke hum apne program me use kar sakte hain.  

**Steps to Create a Custom Module:**  
1. Ek `.py` file create karo (example: `mymodule.py`)  
2. Usme functions, variables ya classes define karo  
3. Dusre program me `import mymodule` karke use karo  

---

```python
# Example 1: Creating a simple module with a function
# File: mymodule1.py
def greet(name):
    return f"Hello, {name}!"
```

```python
greet("Sher Khan")
```

```python

# Example 2: Module with multiple functions
# File: mymodule2.py
def add(a, b):
    return a + b
```

```python
def multiply(a, b):
    return a * b
```

```python
# Example 3: Module with a variable
# File: mymodule3.py
pi = 3.14159
def area_circle(radius):
    return pi * radius ** 2
```

```python

# Example 4: Module with a class
# File: mymodule4.py
class Person:
    def __init__(self, name):
        self.name = name
    
    def greet(self):
        return f"Hello, my name is {self.name}"


```

```python
# Example 5: Module combining functions and variables
# File: mymodule5.py
version = "1.0"

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b
```
