
# 5️⃣ Python Package

**Explanation:**  
Python me **package** ek **folder/directory** hota hai jisme multiple **modules (.py files)** store hote hain.  
- Package ka use **large projects me code organize aur modular** banane ke liye hota hai.  
- Ek package me **`__init__.py`** file hoti hai jo package ko **Python interpreter ke liye recognize** karwati hai.  
  - Python 3.3+ me ye file optional hai, lekin mostly use ki jati hai.  
- Package ke modules ko import karke hum **nested modules** aur **sub-packages** bhi use kar sakte hain.  

**Structure Example:**
- mypackage/
- init.py
- module1.py
- module2.py
- subpackage/
- init.py
- module3.py



# Python package banane ka process step-by-step




## Step 1: Create a Folder for Package

- Jupyter Notebook me **new folder** create karo, jaise `mypackage`.
- Ye folder **package ka root** hoga.




## Step 2: Add `__init__.py` File

- Package folder me **`__init__.py`** file create karo.
- Ye file **empty** ho sakti hai ya initialization code contain kar sakti hai.




## Step 3: Add Modules

- Package folder me multiple **.py files (modules)** create karo.  
  Example:

- mypackage/
- init.py
- module1.py
- module2.py



# Step 4: Import Package and Modules

- Notebook me package import karke modules ke functions use karo.
