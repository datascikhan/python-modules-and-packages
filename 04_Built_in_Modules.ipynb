# 4️⃣ Built-in Modules

Python me **built-in modules** wo modules hote hain jo **Python ke saath pehle se available** hote hain.  
In modules ko import karke hum **complex functionality** easily apne programs me use kar sakte hain bina extra code likhe.  

**Popular Built-in Modules:**

1. **`math`** → Mathematical operations ke liye use hota hai, jaise square root, power, trigonometry, factorial, etc.  
2. **`random`** → Random numbers generate karne aur sequences me random selections karne ke liye use hota hai.  
3. **`os`** → Operating system ke functions ke liye, jaise file handling, directories, environment variables, path operations.  
4. **`sys`** → Python interpreter se related operations ke liye, jaise command-line arguments, system path, exit functions.  
5. **`datetime`** → Date aur time se related operations ke liye, jaise current date/time, formatting, time differences.  

**Benefits of Built-in Modules:**  
- Ready-to-use functionality without extra code  
- Reliable and well-tested  
- Saves development time  
- Supports modular programming and code reuse



## 1️⃣ `math` Module  
`math` module **mathematical operations** provide karta hai jaise square root, power, trigonometry, factorial, logarithms, etc.  


```python

# Example 1: Square root
import math
print(math.sqrt(16)) 
```

```python
# Example 2: Power
import math
print(math.pow(2, 3))
```

```python

# Example 3: Factorial
import math
print(math.factorial(5))
```

```python
# Example 4: Trigonometric function
import math
print(math.sin(math.pi/2)) 
```

```python
# Example 5: Logarithm
import math
print(math.log(100, 10))
```

# 2️⃣ random Module

- random module random numbers generate karne aur sequences se random selections ke liye use hota hai.



```python
# Example 1: Random integer
import random
print(random.randint(1, 10))
```

```python
# Example 2: Random float
import random
print(random.random()) 
```

```python
# Example 3: Random choice from a list
import random
lst = [1, 2, 3, 4, 5]
print(random.choice(lst))
```

```python
# Example 4: Shuffle a list
import random
lst = [1,2,3,4,5]
random.shuffle(lst)
print(lst)
```

```python
# Example 5: Random sample
import random
lst = [1,2,3,4,5]
print(random.sample(lst, 3)) 
```


# 3️⃣ os Module


- os module **operating system related tasks** ke liye use hota hai, jaise file operations, directories, **environment** variables, **path handling**.



```python
# Example 1: Current working directory
import os
print(os.getcwd())
```

```python
# Example 2: List files in a directory
import os
print(os.listdir('.'))
```

```python
# Example 3: Create a new directory
import os
os.mkdir("new_folder")
```

```python
# Example 4: Remove a directory
import os
os.rmdir("new_folder")
```

```python

# Example 5: Check if path exists
import os
print(os.path.exists("example.txt"))
```

# 4️⃣ sys Module

```python

# Example 1: Command-line arguments
import sys
print(sys.argv)  # List of command-line arguments
```

```python
# Example 2: Python version
import sys
print(sys.version)
```

```python

# Example 3: System path
import sys
print(sys.path)
```

```python
# Example 5: Platform info
import sys
print(sys.platform)
```

# 5️⃣ datetime Module

- datetime module date aur time operations ke liye use hota hai, jaise current date/time, formatting, time differences.



```python
# Example 1: Current date and time
from datetime import datetime
now = datetime.now()
print(now)
```

```python

# Example 2: Current date
from datetime import date
today = date.today()
print(today)
```

```python
# Example 3: Format date
from datetime import datetime
now = datetime.now()
print(now.strftime("%d-%m-%Y %H:%M:%S"))
```

```python
# Example 4: Create specific date
from datetime import date
d = date(2025, 10, 30)
print(d)
```

```python
# Example 5: Date difference
from datetime import date
d1 = date(2025, 10, 30)
d2 = date(2025, 1, 1)
diff = d1 - d2
print(diff.days, "days")
```
